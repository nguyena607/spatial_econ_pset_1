```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
housing = pd.read_csv("datasets/CA_housing.csv")
population = pd.read_csv("datasets/CA_pop.csv")
empl_wages = pd.read_csv("datasets/Wages_Emp.csv")
commuting_flows = pd.read_csv("datasets/Commuting_Flows.csv")

socal = ['Los Angeles', 'San Diego', 'Orange', 'Riverside', 'San Bernardino', 'Ventura']
```

```python
#Cleaning the housing data

housing = housing.rename(columns={"           Total Housing": "Total Housing 24", 
                                  "Unnamed: 2": "Total Housing 25",
                                "Percent": "Percent Change",
                                'State/County': 'County'
                                }) 

housing = housing.iloc[2:]
housing = housing[housing['County'].isin(socal)].reset_index(drop=True)

num_cols = [
    "Total Housing 24",
    "Total Housing 25",
    "Percent Change",
]

for c in num_cols:
    housing[c] = (housing[c].astype(str)
                  .str.replace(",", "", regex=False)
                  .str.strip())
    housing[c] = pd.to_numeric(housing[c], errors="coerce")

housing.to_csv("datasets/socal_housing.csv", index=False)
```

```python
#Cleaning the population dataset 

population = population.rename(columns={"           Total Population": "Total Population 24", 
                                  "Unnamed: 2": "Total Population 25",
                                "Percent": "Percent Change",
                                'State/County': 'County'
                                }) 

population = population.iloc[2:]
population = population[population['County'].isin(socal)].reset_index(drop=True)

num_cols = [
    "Total Population 24",
    "Total Population 25",
    "Percent Change",
]

for c in num_cols:
    population[c] = (population[c].astype(str)
                  .str.replace(",", "", regex=False)
                  .str.strip())
    population[c] = pd.to_numeric(population[c], errors="coerce")

population.to_csv("datasets/socal_population.csv", index=False)
```

```python
#Cleaning the employment and wages dataset 

empl_wages = empl_wages.rename(columns={"Annual\nEstablishments": "Num_Establishments", 
                                  "Annual\nAverage\nEmployment": "Employment",
                                "Total\nAnnual\nWages": "Annual_Wages",
                                "Annual\nAverage\nWeekly Wage": "Weekly_Wage",
                                "Annual\nWages per\nEmployee": "Wage_per_Employee",
                                })
empl_wages.head()

empl_wages = empl_wages.iloc[:, :6]


empl_wages["County"] = empl_wages["County"].str.replace(r"\s*County,\s*California\s*$", "", regex=True)
empl_wages = empl_wages[empl_wages['County'].isin(socal)].reset_index(drop=True)

we = empl_wages.copy()   

num_cols = [
    "Num_Establishments",
    "Employment",           
    "Annual_Wages",
    "Weekly_Wage",
    "Wage_per_Employee",
]

#converting to numeric 
for c in num_cols:
    we[c] = (we[c].astype(str)
                  .str.replace(",", "", regex=False)
                  .str.strip())
    we[c] = pd.to_numeric(we[c], errors="coerce")

we.to_csv("datasets/socal_empl_wages.csv", index=False)
```

```python
commuting_flows_ = commuting_flows[(commuting_flows['State Name'] == 'California')]
ca_to_ca_commuting = commuting_flows_[commuting_flows_['State Name.1'] == 'California']
ca_to_ca_commuting = commuting_flows_[commuting_flows_['State Name.1'] == 'California']
ca_to_ca_commuting = ca_to_ca_commuting[['State Name', 'County Name', 'State Name.1', 'County Name.1', 'Workers in Commuting Flow', 'Margin of Error']]
ca_to_ca_commuting = ca_to_ca_commuting.rename(columns={'County Name': 'Residence_County', 'County Name.1': 'Workplace_County'})    
ca_to_ca_commuting["Residence_County"] = ca_to_ca_commuting["Residence_County"].str.rsplit(" ", n=1).str[0]
ca_to_ca_commuting["Workplace_County"] = ca_to_ca_commuting["Workplace_County"].str.rsplit(" ", n=1).str[0]

ca_to_ca_commuting = ca_to_ca_commuting[ca_to_ca_commuting['Residence_County'].isin(socal) & ca_to_ca_commuting['Workplace_County'].isin(socal)]

ca_to_ca_commuting.to_csv("datasets/socal_commuting_flows.csv", index=False)
```
